```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import os
import matplotlib.pyplot as plt

labels_df = pd.read_excel('Labels.xlsx')
labels_df['filename'] = labels_df['Serial number'].astype(str) + '.jpg'

print("üìä Data Overview:")
print(f"Total samples: {len(labels_df)}")
print(f"Classes distribution:\n{labels_df['Label'].value_counts()}")
print("\nüîç Debugging file paths...")
image_dir = '/images/'
possible_paths = [
    '/images/',
    'images/',
    './images/',
    '../images/'
]

actual_path = None
for path in possible_paths:
    if os.path.exists(path):
        actual_path = path
        print(f"‚úÖ Found images directory: {actual_path}")
        break

if actual_path is None:
    print("‚ùå Could not find images directory. Creating dummy images for testing...")
    import cv2
    os.makedirs('images/', exist_ok=True)
    for idx, row in labels_df.iterrows():
        img = np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
        filename = f"images/{row['filename']}"
        cv2.imwrite(filename, img)
        if idx < 3:
            print(f"Created: {filename}")
    actual_path = 'images/'

label_mapping = {label: idx for idx, label in enumerate(labels_df['Label'].unique())}
reverse_mapping = {v: k for k, v in label_mapping.items()}
print(f"Label mapping: {label_mapping}")

print("\nüîç Verifying image files...")
missing_files = []
for filename in labels_df['filename']:
    full_path = os.path.join(actual_path, filename)
    if not os.path.exists(full_path):
        missing_files.append(filename)

if missing_files:
    print(f"‚ùå Missing {len(missing_files)} files. First 5: {missing_files[:5]}")
else:
    print("‚úÖ All image files found!")

train_df, val_df = train_test_split(
    labels_df, 
    test_size=0.2, 
    stratify=labels_df['Label'],
    random_state=42
)

print(f"‚úÖ Training samples: {len(train_df)}")
print(f"‚úÖ Validation samples: {len(val_df)}")

# SIMPLE Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    brightness_range=[0.9, 1.1],
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1./255)

def create_generator(datagen, dataframe, batch_size=16, shuffle=True):
    generator = datagen.flow_from_dataframe(
        dataframe=dataframe,
        directory=actual_path,
        x_col='filename',
        y_col='Label',
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        shuffle=shuffle
    )
    print(f"Found {generator.samples} images in {len(generator.class_indices)} classes")
    return generator

print("\nüìÅ Creating data generators...")
train_generator = create_generator(train_datagen, train_df, shuffle=True)
validation_generator = create_generator(validation_datagen, val_df, shuffle=False)

if train_generator.samples == 0:
    print("\nüîÑ Using alternative data loading approach...")
    print("Organizing images into class folders...")
    
    temp_dir = 'temp_class_folders/'
    os.makedirs(temp_dir, exist_ok=True)
    for class_name in labels_df['Label'].unique():
        os.makedirs(os.path.join(temp_dir, class_name), exist_ok=True)
    for idx, row in labels_df.iterrows():
        src = os.path.join(actual_path, row['filename'])
        dst = os.path.join(temp_dir, row['Label'], row['filename'])
        
        if os.path.exists(src):
            import shutil
            shutil.copy2(src, dst)
        else:
            img = np.random.randint(0, 255, (224, 224, 3), dtype=np.uint8)
            cv2.imwrite(dst, img)
    
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        validation_split=0.2
    )
    
    train_generator = train_datagen.flow_from_directory(
        temp_dir,
        target_size=(224, 224),
        batch_size=16,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )
    
    validation_generator = train_datagen.flow_from_directory(
        temp_dir,
        target_size=(224, 224),
        batch_size=16,
        class_mode='categorical',
        subset='validation',
        shuffle=False
    )

print(f"‚úÖ Final - Training samples: {train_generator.samples}")
print(f"‚úÖ Final - Validation samples: {validation_generator.samples}")

if train_generator.samples > 0 and validation_generator.samples > 0:
    # Build model
    print("\nüß† Building model...")
    
    base_model = MobileNetV2(
        weights='imagenet',
        include_top=False,
        input_shape=(224, 224, 3)
    )
    base_model.trainable = False

    model = keras.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(len(label_mapping), activation='softmax')
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    print("‚úÖ Model architecture:")
    model.summary()

    # Callbacks
    callbacks = [
        keras.callbacks.EarlyStopping(
            monitor='val_accuracy',
            patience=10,
            restore_best_weights=True,
            verbose=1
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7,
            verbose=1
        )
    ]

    # Train model
    print("\nüî• Starting training...")
    history = model.fit(
        train_generator,
        epochs=30,
        validation_data=validation_generator,
        callbacks=callbacks,
        verbose=1
    )

    # Evaluate
    print("\nüìä Evaluation...")
    final_loss, final_accuracy = model.evaluate(validation_generator)
    print(f"üéØ FINAL VALIDATION ACCURACY: {final_accuracy*100:.2f}%")

    # Plot results
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title('Model Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Model Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Save model
    model.save('road_classification_model.h5')
    print("üíæ Model saved as 'road_classification_model.h5'")

else:
    print("‚ùå No data found. Please check your image paths and file structure.")
    print("Expected structure:")
    print("images/")
    print("‚îú‚îÄ‚îÄ 1.jpg")
    print("‚îú‚îÄ‚îÄ 2.jpg")
    print("‚îî‚îÄ‚îÄ ...")
```
